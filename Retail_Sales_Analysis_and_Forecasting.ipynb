```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Создадим вымышленные данные о продажах
data = {
    'Дата': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Продажи': [15000, 16000, 18000, 19000, 22000, 25000, 28000, 30000, 32000, 34000, 36000, 38000],
    'Рекламный_бюджет': [1000, 1200, 1500, 2000, 2500, 2800, 3000, 3200, 3500, 3800, 4000, 4200]
}

df = pd.DataFrame(data)

# Проведем EDA
print("Статистика по данным:")
print(df.describe())

# Визуализируем продажи и рекламный бюджет
plt.figure(figsize=(10, 5))
plt.plot(df['Дата'], df['Продажи'], marker='o', label='Продажи')
plt.plot(df['Дата'], df['Рекламный_бюджет'], marker='x', label='Рекламный бюджет')
plt.title('Динамика продаж и рекламного бюджета')
plt.xlabel('Дата')
plt.ylabel('Сумма')
plt.legend()
plt.grid()
plt.show()

# Моделирование продаж с использованием линейной регрессии
X = df['Рекламный_бюджет'].values.reshape(-1, 1)
y = df['Продажи'].values
model = LinearRegression()
model.fit(X, y)

# Прогнозирование продаж
future_budget = np.array([4500, 4800, 5000]).reshape(-1, 1)
future_sales = model.predict(future_budget)

print("Прогноз продаж для рекламного бюджета [4500, 4800, 5000]:")
for budget, sales in zip([4500, 4800, 5000], future_sales):
    print(f"Рекламный бюджет: {budget}, Прогноз продаж: {sales:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Создадим вымышленные данные о продажах
data = {
    'Дата': pd.date_range(start='2023-01-01', periods=12, freq='M'),
    'Продажи': [15000, 16000, 18000, 19000, 22000, 25000, 28000, 30000, 32000, 34000, 36000, 38000],
    'Рекламный_бюджет': [1000, 1200, 1500, 2000, 2500, 2800, 3000, 3200, 3500, 3800, 4000, 4200]
}

df = pd.DataFrame(data)

# Проведем EDA
print("Статистика по данным:")
print(df.describe())

# Визуализируем продажи и рекламный бюджет
plt.figure(figsize=(10, 5))
plt.plot(df['Дата'], df['Продажи'], marker='o', label='Продажи')
plt.plot(df['Дата'], df['Рекламный_бюджет'], marker='x', label='Рекламный бюджет')
plt.title('Динамика продаж и рекламного бюджета')
plt.xlabel('Дата')
plt.ylabel('Сумма')
plt.legend()
plt.grid()
plt.show()

# Создадим тепловую карту корреляции
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Корреляция между признаками')
plt.show()

# Моделирование продаж с использованием линейной регрессии
X = df['Рекламный_бюджет'].values.reshape(-1, 1)
y = df['Продажи'].values
model = LinearRegression()
model.fit(X, y)

# Прогнозирование продаж
future_budget = np.array([4500, 4800, 5000]).reshape(-1, 1)
future_sales = model.predict(future_budget)

print("Прогноз продаж для рекламного бюджета [4500, 4800, 5000]:")
for budget, sales in zip([4500, 4800, 5000], future_sales):
    print(f"Рекламный бюджет: {budget}, Прогноз продаж: {sales:.2f}")

# Создадим таблицу с данными
plt.figure(figsize=(8, 4))
plt.axis('off')
plt.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
plt.title('Данные о продажах и рекламном бюджете')
plt.show()
```
